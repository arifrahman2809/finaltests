```python
url = "https://berita-indo-api-next.vercel.app/api/cnn-news/teknologi"
response = requests.get(url)

# Pastikan responsenya berhasil (status code 200)
if response.status_code == 200:
    json_data = response.json()
else:
    print("Gagal mendapatkan data dari API")
```

```python
if 'data' in json_data:
    data = json_data['data']
    df = pd.DataFrame(data)
else:
    print("Key 'data' tidak ditemukan dalam JSON")
```

```python
df['isoDate'] = pd.to_datetime(df['isoDate'])
```

```python
aggregated_df = df.groupby('creator').size().reset_index(name='jumlah_berita')
sorted_df = aggregated_df.sort_values(by='jumlah_berita', ascending=False)
```
